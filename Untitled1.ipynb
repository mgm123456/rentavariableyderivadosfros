```python
import random
import matplotlib.pyplot as plt
random.seed()
#Valor Inicial de la acción
Accion = 1000     
 #Número de periodos considerados para el paseo aleatorio
n = 30            
#Variación
factor = 1.03       
#Probabilidad otorgada a cada opción (aumento o disminución), 50-50
p=0.5            
Precio = [Accion] 
for t in range(1,n):
    if random.random() < p:
        Accion *= factor
    else:
        Accion /= factor
    Precio.append(Accion)

#Generamos el gráfico del paseo aleatorio en una dimensión
plt.plot(range(n),Precio)
plt.show()               
Varianza= (Accion/999) - 1
#Calculamos la variación y la incluimos debajo del gráfico
print(f"La variación de la serie aleatoria es {Varianza:.3%} ")  
```
